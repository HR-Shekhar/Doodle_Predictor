```python
import numpy as np
import matplotlib.pyplot as plt

data = np.load('./raw_Data/t-shirt.npy')  # shape: (N, 784) or (N, 28, 28)
print(data.shape)
```

```python
images = data.reshape((-1, 28, 28))  # Now shape is (N, 28, 28)
```

```python
fig, axs = plt.subplots(8, 5, figsize=(10, 10))
for i in range(40):
    ax = axs[i // 5, i % 5]
    ax.imshow(images[i], cmap='gray')
    ax.axis('off')
plt.tight_layout()
plt.show()
```

```python
np.random.seed(42)  # Optional but important — for reproducibility
random_indices = np.random.choice(len(data), 30000, replace=False)

# Step 3: Extract those samples
subset = data[random_indices]
```

```python
np.save("./data/t-shirt.npy", subset)
```

```python
image_samp = subset.reshape((-1, 28, 28))  # Now shape is (N, 28, 28)
```

```python
fig, axs = plt.subplots(8, 5, figsize=(10, 10))
for i in range(40):
    ax = axs[i // 5, i % 5]
    ax.imshow(image_samp[i], cmap='gray')
    ax.axis('off')
plt.tight_layout()
plt.show()
```
